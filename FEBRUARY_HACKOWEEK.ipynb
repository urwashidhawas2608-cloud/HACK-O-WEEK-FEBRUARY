```python
import spacy
import re
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
def extract_entities(question):

    semester = None
    course = None
    date = None

    # Regex patterns
    sem_pattern = r"SEM\s?\d"
    course_pattern = r"\b(CS|AIML|DS|IT|ME|CE)\b"

    # Extract semester
    sem_match = re.search(sem_pattern, question.upper())
    if sem_match:
        semester = sem_match.group()

    # Extract course
    course_match = re.search(course_pattern, question.upper())
    if course_match:
        course = course_match.group()

    # Extract date using NLP
    doc = nlp(question)
    for ent in doc.ents:
        if ent.label_ == "DATE":
            date = ent.text

    return {
        "Semester": semester,
        "Course": course,
        "Date": date
    }
```

```python
question = "When is SEM 5 CS exam on 15 March?"

entities = extract_entities(question)

print("Extracted Entities:")
print(entities)
```

```python
def chatbot_response(question):

    entities = extract_entities(question)

    semester = entities["Semester"]
    course = entities["Course"]

    if semester and course:
        return f"The exam for {course} in {semester} will be announced soon."

    elif semester:
        return f"The exam schedule for {semester} will be updated on the portal."

    else:
        return "Please specify semester or course."
```

```python
question = input("Ask your question: ")

response = chatbot_response(question)

print("Chatbot:", response)
```

```python
questions = [
    "When is SEM 5 CS exam?",
    "Tell me exam date for SEM 3 AIML",
    "When is DS exam on 10 April?"
]

for q in questions:
    print("Question:", q)
    print("Entities:", extract_entities(q))
    print("Response:", chatbot_response(q))
    print()
```

```python
import spacy
import re
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
def extract_entities(question):

    semester = None
    course = None

    sem_pattern = r"SEM\s?\d|first year|second year|third year|fourth year"
    course_pattern = r"\b(CS|AIML|DS|IT|ME|CE)\b"

    sem_match = re.search(sem_pattern, question.lower())
    if sem_match:
        semester = sem_match.group()

    course_match = re.search(course_pattern, question.upper())
    if course_match:
        course = course_match.group()

    return {
        "Semester": semester,
        "Course": course
    }
```

```python
context = {
    "semester": None,
    "course": None
}
```

```python
supported_keywords = ["exam", "schedule", "date", "semester", "course"]
```

```python
def fallback_response():

    return """
Sorry, I could not understand your question.

You can ask things like:
• When is SEM 5 exam?
• What is the exam date for AIML?
• Show exam schedule for SEM 3

If your issue is different, please contact the academic advisor:
Email: advisor@university.edu
"""
```

```python
def chatbot_response(question):

    global context

    # Check if question is related to supported topics
    if not any(word in question.lower() for word in supported_keywords):
        return fallback_response()

    entities = extract_entities(question)

    if entities["Semester"]:
        context["semester"] = entities["Semester"]

    if entities["Course"]:
        context["course"] = entities["Course"]

    if context["semester"] and context["course"]:
        return f"The exam for {context['course']} in {context['semester']} will be announced soon."

    elif context["semester"]:
        return f"Please specify the course for {context['semester']}."

    elif context["course"]:
        return f"Please specify the semester for {context['course']}."

    else:
        return "Please mention semester or course."
```
